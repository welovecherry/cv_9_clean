```python
# 정답이 있는 샘플만 필터링
df_labeled = df_simple[df_simple['눈 정답'].notna()].copy()

# 데이터 타입을 정수형으로 통일
df_labeled['눈 정답'] = df_labeled['눈 정답'].astype(int)
df_labeled['1_최고점_conv'] = df_labeled['1_최고점_conv'].astype(int)
df_labeled['2_auto_eff_x'] = df_labeled['2_auto_eff_x'].astype(int)
df_labeled['3_tf_eff'] = df_labeled['3_tf_eff'].astype(int)

# 클래스별 정답률 계산 함수
def classwise_accuracy(df, label_col, pred_cols):
    result = {}
    unique_labels = sorted(df[label_col].unique())

    for label in unique_labels:
        df_class = df[df[label_col] == label]
        result[label] = {
            col: round((df_class[col] == df_class[label_col]).mean() * 100, 1)
            for col in pred_cols
        }

    return pd.DataFrame(result).T

# 예측 칼럼
pred_cols = ['1_최고점_conv', '2_auto_eff_x', '3_tf_eff']

# 정답률 계산 및 출력
class_accuracy_df = classwise_accuracy(df_labeled, '눈 정답', pred_cols)
print("클래스별 정답률 (%):")
print(class_accuracy_df)
```

•	클래스 14:

	•	2_auto_eff_x는 특히 13을 5번 예측
	•	3_tf_eff는 10을 5번 예측 → 많은 클래스와의 혼동 유발

•	클래스 12:
	•	1_최고점_conv: 3, 12, 14로 다양하게 예측
	•	3_tf_eff: 6, 10 등으로 분산
 
•	클래스 3:
	•	각 모델에서 3, 4, 6, 7, 10, 13으로 분산 → 뚜렷한 경향 없음

```python
from collections import defaultdict, Counter

def model_prediction_distribution_by_class(df, label_col, pred_cols):
    class_model_preds = defaultdict(lambda: defaultdict(list))

    # 각 행의 눈 정답 기준으로 모델별 예측값 저장
    for _, row in df.dropna(subset=[label_col]).iterrows():
        true_label = int(row[label_col])
        for model_col in pred_cols:
            class_model_preds[true_label][model_col].append(int(row[model_col]))

    # 각 클래스별 모델 예측 분포를 정리
    rows = []
    for cls in sorted(class_model_preds.keys()):
        row = {'눈 정답 클래스': cls}
        for model_col in pred_cols:
            preds = class_model_preds[cls][model_col]
            pred_counts = Counter(preds)
            # 예: "3(5), 12(2)" 형식으로 정리
            summary = ', '.join([f'{k}({v})' for k, v in sorted(pred_counts.items())])
            row[f'{model_col} 분포'] = summary
        rows.append(row)

    return pd.DataFrame(rows)

# 실행
pred_cols = ['1_최고점_conv', '2_auto_eff_x', '3_tf_eff']
distribution_df = model_prediction_distribution_by_class(df_simple, '눈 정답', pred_cols)
distribution_df
```
